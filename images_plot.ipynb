```python
import matplotlib.pyplot as plt
import tifffile
import numpy as np
from glob import glob

#  
img = tifffile.imread("images/EGST39_A_Image 2_30min.tif")

#     (:   )
t = 100
img_t = img[t]

# z-  
fig, axes = plt.subplots(1, img_t.shape[0], figsize=(100, 100))
for i in range(img_t.shape[0]):
    axes[i].imshow(img_t[i], cmap='gray')
    axes[i].axis('off')
    axes[i].set_title(f'z = {i}')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming img_t is the 3D image array with shape (z_slices, height, width)
z_slices = img_t.shape[0]
rows = (z_slices + 3) // 4  # Calculate the number of rows needed

fig, axes = plt.subplots(rows, 4, figsize=(20, rows * 5))

for i in range(z_slices):
    row = i // 4
    col = i % 4
    axes[row, col].imshow(img_t[i], cmap='gray')
    axes[row, col].axis('off')
    axes[row, col].set_title(f'z = {i}')

# Turn off axes for any empty subplots
for j in range(z_slices, rows * 4):
    row = j // 4
    col = j % 4
    axes[row, col].axis('off')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

t = 0
img_t = img[t]
max_projection = np.max(img_t, axis=0)

plt.imshow(max_projection, cmap='gray')
plt.title('Maximum Intensity Projection')
plt.axis('off')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

# 3D 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

#     (:   )
t = 0
img_t = img[t]

#  
x, y, z = np.indices(img_t.shape)

#   
voxels = img_t > 0

# 3D 
ax.voxels(voxels, facecolors='blue', edgecolor='k')

plt.show()
```
